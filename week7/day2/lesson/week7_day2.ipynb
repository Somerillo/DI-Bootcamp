#Google Search and input text

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Initialize WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=options)

try:
    # Open Google
    driver.get("https://www.google.com")

    # Find the search bar and perform a search
    search_box = driver.find_element(By.NAME, "q")
    search_box.send_keys("Selenium Python")
    search_box.send_keys(Keys.RETURN)

    # Wait for results to load
    time.sleep(2)

    # Print the titles of the top search results
    results = driver.find_elements(By.XPATH, '//h3')
    print("Top Results:")
    for result in results[:5]:  # Print the top 5 results
        print(result.text)

except Exception as e:
    print(f"An error occurred: {e}")

time.sleep(10)

````

#Login to Heroku and Submit

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=options)

try:
    # Open the demo login page
    driver.get("https://the-internet.herokuapp.com/login")

    # Enter credentials
    driver.find_element(By.ID, "username").send_keys("tomsmith")
    driver.find_element(By.ID, "password").send_keys("SuperSecretPassword!")
    time.sleep(2)
    # Click the login button
    driver.find_element(By.XPATH, '//button[@type="submit"]').click()

    # Wait and verify the login
    time.sleep(2)
    message = driver.find_element(By.ID, "flash").text
    print("Login Message:", message)

finally:
    # Close the browser
    driver.quit()
````

#Scrape all quotes from website

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=options)

try:
    # Open the quotes page
    driver.get("http://quotes.toscrape.com/")

    # Scrape quotes and authors
    quotes = driver.find_elements(By.CLASS_NAME, "quote")
    for quote in quotes:
        text = quote.find_element(By.CLASS_NAME, "text").text
        author = quote.find_element(By.CLASS_NAME, "author").text
        print(f"{text} - {author}")

finally:
    # Close the browser
    driver.quit()
````

#Scroll on Website to find link and press it

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")  # Open browser in maximized mode
driver = webdriver.Chrome(options=options)

try:
    # Open the infinite scrolling page
    driver.get("https://infinite-scroll.com/demo/full-page/")

    # Scroll to the bottom multiple times to load content
    for _ in range(5):  # Adjust the range as needed to ensure the element loads
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Wait for new content to load

    # Locate the link by its text or href attribute
    target_link = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//a[contains(@href, 'https://codepen.io/desandro/pens/tags/?grid_type=list&selected_tag=packery-docs')]"))
    )

    # Click the element
    target_link.click()
    print("Successfully clicked on the 'CodePens for every feature' link!")

    # Optionally, wait to observe the result
    time.sleep(5)

finally:
    # Close the browser
    driver.quit()
````

Here's a **Selenium Cheatsheet** summarizing all the concepts, methods, and techniques we've used in this lesson, organized by categories:

---

### **1. Basic Setup**
- **Installing Selenium**:
  ```bash
  pip install selenium
  ```

- **Initializing WebDriver**:
  ```python
  from selenium import webdriver
  driver = webdriver.Chrome(options=options)
  ```

- **Chrome Options**:
  ```python
  from selenium.webdriver.chrome.options import Options
  options = Options()
  options.add_argument("--start-maximized")   # Open browser in maximized mode
  options.add_argument("--headless")         # Run in headless mode
  options.add_argument("--no-sandbox")       # Bypass OS-level security model
  options.add_argument("--disable-dev-shm-usage")  # Prevent resource errors
  driver = webdriver.Chrome(options=options)
  ```

---

### **2. Navigating Web Pages**
- **Open a URL**:
  ```python
  driver.get("https://example.com")
  ```

- **Get the Current Page Title**:
  ```python
  print(driver.title)
  ```

- **Get the Current Page URL**:
  ```python
  print(driver.current_url)
  ```

---

### **3. Finding Elements**
- **By ID**:
  ```python
  driver.find_element(By.ID, "element-id")
  ```

- **By Class Name**:
  ```python
  driver.find_element(By.CLASS_NAME, "class-name")
  ```

- **By Name**:
  ```python
  driver.find_element(By.NAME, "element-name")
  ```

- **By Tag Name**:
  ```python
  driver.find_elements(By.TAG_NAME, "a")  # Find all <a> tags
  ```

- **By XPath**:
  ```python
  driver.find_element(By.XPATH, "//tagname[@attribute='value']")
  ```

- **By Link Text**:
  ```python
  driver.find_element(By.LINK_TEXT, "Exact Link Text")
  ```

- **By Partial Link Text**:
  ```python
  driver.find_element(By.PARTIAL_LINK_TEXT, "Partial Text")
  ```

---

### **4. Interacting with Elements**
- **Send Text to an Input Field**:
  ```python
  element = driver.find_element(By.ID, "username")
  element.send_keys("my_username")
  ```

- **Click an Element**:
  ```python
  button = driver.find_element(By.ID, "submit-button")
  button.click()
  ```

- **Clear Text from an Input Field**:
  ```python
  element.clear()
  ```

### **7. Explicit Waits**
- **Wait for an Element to Be Clickable**:

### **9. XPath Examples**
| **XPath Expression**                          | **Description**                                              |
|-----------------------------------------------|--------------------------------------------------------------|
| `//tagname`                                   | Finds all elements with the given tag name.                  |
| `//tagname[@attr='value']`                    | Finds elements with the specified attribute.                 |
| `//*[contains(@attr, 'partial_value')]`       | Finds elements where an attribute contains a substring.      |
| `//tagname[text()='Exact Text']`              | Finds elements containing the exact text.                    |
| `//tagname[contains(text(), 'Partial Text')]` | Finds elements containing partial text.                      |
| `//tagname[@attr='value' and @attr2='value2']`| Combines multiple conditions using `and`.                    |

````python
wait = WebDriverWait(driver, 10)
# Extract quotes details
quotes_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.quote')))

quotes_list = []
for quote_elem in quotes_elements:
    text = quote_elem.find_element(By.CLASS_NAME, 'text').text
    author = quote_elem.find_element(By.CLASS_NAME, 'author').text
    tags = quote_elem.find_elements(By.CLASS_NAME, 'tag')
    tag_list = [tag.text for tag in tags]

    quotes_list.append({
        'author': author,
        'text': text,
        'tags': tag_list,
    })

pprint.pprint(quotes_list)
````

````python
# Step 1: Install dependencies
# !apt-get update
# !apt-get install -y chromium-browser chromium-chromedriver
# !pip install selenium

# Step 2: Import necessary libraries and configure options
from selenium import webdriver
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run Chrome in headless mode
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Step 3: Define example pages as inline HTML
example_pages = {
    "login.html": """
    <!DOCTYPE html>
    <html>
    <body>
        <h1>Login Page</h1>
        <form>
            <label for="username">Username:</label><br>
            <input type="text" id="username" name="username"><br>
            <label for="password">Password:</label><br>
            <input type="password" id="password" name="password"><br><br>
            <button type="button" id="login">Login</button>
        </form>
    </body>
    </html>
    """,
    "search.html": """
    <!DOCTYPE html>
    <html>
    <body>
        <h1>Search Page</h1>
        <form>
            <label for="search">Search:</label><br>
            <input type="text" id="search" name="search" placeholder="Type something..."><br><br>
            <button type="button" id="search-btn">Search</button>
        </form>
    </body>
    </html>
    """,
    "counter.html": """
    <!DOCTYPE html>
    <html>
    <body>
        <h1>Click Counter</h1>
        <button id="counter-btn">Click me!</button>
        <p id="counter">Count: 0</p>
        <script>
            let count = 0;
            document.getElementById('counter-btn').onclick = function() {
                count++;
                document.getElementById('counter').innerHTML = 'Count: ' + count;
            };
        </script>
    </body>
    </html>
    """,
}

# Step 4: Save pages as local HTML files
import os

os.makedirs("selenium_examples", exist_ok=True)
for filename, content in example_pages.items():
    with open(f"selenium_examples/{filename}", "w") as f:
        f.write(content)

# Step 5: Serve the pages locally using a temporary HTTP server
import threading
from http.server import HTTPServer, SimpleHTTPRequestHandler

os.chdir("selenium_examples")
server = HTTPServer(("localhost", 8000), SimpleHTTPRequestHandler)
thread = threading.Thread(target=server.serve_forever, daemon=True)
thread.start()

# Step 6: Automate interactions with Selenium
try:
    driver = webdriver.Chrome(options=options)

    # Example 1: Interact with login page
    driver.get("http://localhost:8000/login.html")
    driver.find_element(By.ID, "username").send_keys("test_user")
    driver.find_element(By.ID, "password").send_keys("secure_password")
    print("Login page interacted successfully.")

    # Example 2: Perform a search
    driver.get("http://localhost:8000/search.html")
    search_box = driver.find_element(By.ID, "search")
    search_box.send_keys("Selenium Python")
    driver.find_element(By.ID, "search-btn").click()
    print("Search page interacted successfully.")

    # Example 3: Increment the counter
    driver.get("http://localhost:8000/counter.html")
    button = driver.find_element(By.ID, "counter-btn")
    for _ in range(3):  # Click 3 times
        button.click()
    counter_text = driver.find_element(By.ID, "counter").text
    print("Counter text after clicks:", counter_text)

finally:
    driver.quit()
    server.shutdown()
    print("Server shut down.")
````
