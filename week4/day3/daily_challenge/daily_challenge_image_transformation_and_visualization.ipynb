# To rotate **one image** 90 degrees:

```python
from PIL import Image

# define the path to the image
path = '/content/flowers/flowers/01_031.png'

# load the image
image = Image.open(path)

# display the image inline
plt.imshow(image)
plt.show()
```

```python
def rotate_image(image, angle):
  return rotate(image, angle=angle, axes=(1, 0), reshape=True)

image_rotated = rotate_image(image, 90)

# display the image inline
plt.imshow(image_rotated)
plt.show()
```

# Flip an image horizontally and then vertically:

```python
from PIL import Image, ImageOps

def flip_image(image, direction='vertical'):
    if direction == 'vertical':
        return ImageOps.flip(image)  # Flip vertically
    elif direction == 'horizontal':
        return ImageOps.mirror(image)  # Flip horizontally
    else:
        raise ValueError("FTS!!!!!!")

image_flipped = flip_image(image, 'horizontal')

# display the image inline
plt.imshow(image_flipped)
plt.show()
```

```python
image_flipped = flip_image(image_flipped, 'vertical')

# display the image inline
plt.imshow(image_flipped)
plt.show()
```

# Zoom in on an image (scale by 1.2x) using .resize(...)

```python
def zoom_image(image, zoom_=1.2):

  # calculate new dimensions
  new_width = int(image.width * zoom_)
  new_height = int(image.height * zoom_)

  # resizing
  zoomed_image = image.resize((new_width, new_height), resample=Image.LANCZOS)

  return zoomed_image

image_zoomed = zoom_image(image, 1.2)
# display the image inline
plt.imshow(image_zoomed)
plt.show()
```

# Display the original and augmented images side by side

```python
# list the images
list_img = [image, image_zoomed]
titles = ["original", "zoomed 1.2x"]

# set figure size to plot
plt.figure(figsize=(10, 5))

# Loop through the list of images and display them
for i, img in enumerate(list_img):
  # subplot for each image
  ax = plt.subplot(1, 2, i + 1)

  # title for each subplot
  ax.title.set_text(titles[i])

  # display the image
  plt.imshow(img)

plt.show()
```
