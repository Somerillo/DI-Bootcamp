## ðŸŒŸ Exercise 5 : Export a dataframe to excel format and JSON format.

    Create a simple dataframe.
    Export the dataframe to an excel file.
    Export the dataframe to a JSON file.


```python
from faker import Faker

fake = Faker()

rows = []
for _ in range(15):
    data = {
        'name': fake.first_name(),
        'email': fake.last_name(),
        'genre': fake.email(),
        'birthdate': fake.date_of_birth(minimum_age=18, maximum_age=99)
    }
    rows.append(data)

df = pd.DataFrame(rows)
df.head()
```

```python
# export to excel
df.to_excel('simple_dataframe.xlsx', index=False) # we dont need the index

# export to json
df.to_json('simple_dataframe.json', orient='records') # each row of the DataFrame is converted into a separate JSON object
```
