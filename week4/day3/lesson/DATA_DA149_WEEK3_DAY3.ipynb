```python
# Sample JSON-like data
data = {
    "user_id": 12345,
    "name": {"first": "john", "last": "doe"},
    "orders": [
        {
            "order_id": "Abc123",
            "date": "2023-01-15",
            "items": [
                {"product": "Laptop", "price": 100, "quantity": 1},
                {"product": "Iphone", "price": 200, "quantity": 1}
            ],
        },
        {
            "order_id": "def123",
            "date": "2023-01-15",
            "items": [
                {"product": "Laptop", "price": 100, "quantity": 1},
                {"product": "Iphone", "price": 100, "quantity": 1}
            ],
        },
    ],
}

# Target order_id (case-insensitive)
target_order_id = "Abc123"

# Find the target order
target_order = next(
    (order for order in data["orders"] if order["order_id"].lower() == target_order_id.lower()),
    None
)

# Extract the price of the second product if the order exists
if target_order and len(target_order["items"]) > 1:
    second_product_price = target_order["items"][1]["price"]
    print(f"The price of the second product is: {second_product_price}")
else:
    print("Order or second product not found.")
```

```python
# Sample data
data = {
    "user_id": 12345,
    "name": {"first": "john", "last": "doe"},
    "orders": [
        {
            "order_id": "Abc123",
            "date": "2023-01-15",
            "items": [
                {"product": "Laptop", "price": 100, "quantity": 1},
                {"product": "Iphone", "price": "$200", "quantity": 1}
            ],
        },
        {
            "order_id": "def123",
            "date": "2023-01-15",
            "items": [
                {"product": "Laptop", "price": 100, "quantity": 1},
                {"product": "Iphone", "price": 100, "quantity": 1}
            ],
        },
    ],
}

# Target order_id (case-insensitive)
target_order_id = "Abc123"

# Find the order and get the second product price
try:
  for order in data["orders"]:
      if order["order_id"].lower() == target_order_id.lower():
          second_product_price = int(order["items"][1]["price"])
          print(f"The price of the second product is: {second_product_price}")
          break
  else:
      print("Order not found or second product is missing.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
# to randomly rotate the image by an angle between -20 to +20 degrees.
from tensorflow.keras.preprocessing.image import ImageDataGenerator
data_generator = ImageDataGenerator(rotation_range=20)

# to rotate the image buy exactly 30 degrees
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
# Rotate image by 30 degrees
  return rotate(image, 30, reshape=False, mode='nearest')
#  Create ImageDataGenerator with custom preprocessing function
data_generator = ImageDataGenerator(preprocessing_function=rotate_image_30_degrees)
```

```python
image_iterator = data_generator.flow_from_directory("/content", target_size=(224, 224))
```

```python
import matplotlib.pyplot as plt

# Get a batch of augmented images
batch_images, batch_labels = next(image_iterator)

# Plot the first 5 augmented images
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(batch_images[i].astype('uint8'))
    plt.axis('off')
plt.show()
```
