```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ds_salaries.csv', index_col=0) # index col = 0
df.head()
```

Quantitative (numerically measurable) columns are `[work_year, salary, salary_in_usd]`. The rest is quantitative, although `company_size` and `experience_level` coluld be treated as quantitative depending on the analysis.

## Preprocess the data by handling missing values and removing duplicates if needed.

```python
df.info()
```

No missing values after applying `info` method.

Then we proceed to remove duplicates:

```python
print(f'there are {df.duplicated().sum()} duplicates')

df = df.drop_duplicates()

print(f'after removing them we verify that now we have {df.duplicated().sum()} duplicated rows')
```

```python
df.describe()
```

Notice the lowest salary is 4k USD

```python
# # we keep only relevant columns for our analysis
# df = df[['experience_level', 'salary_in_usd']]
```

Outliers removal:

```python
# for outlier removal we employ IQR
Q1 = df['salary_in_usd'].quantile(0.25)
Q3 = df['salary_in_usd'].quantile(0.75)
IQR = Q3 - Q1

# define limits
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

df_cleaned = df[(df['salary_in_usd'] >= lower) & (df['salary_in_usd'] <= upper)]
```

## Perform a group-wise analysis of the `experience_level` column, calculating the mean and median salaries for each group using Pandas’ .groupby.

```python
# group by 'experience_level' then aggregates
grouped_salaries = df_cleaned.groupby('experience_level')['salary_in_usd'].agg(['mean', 'median']).reset_index() # reset_index to plot!!!!!!!!
grouped_salaries = grouped_salaries.sort_values(by='mean')

grouped_salaries
```

The high tech industry used to have very nice entrance salaries at the beggining

## Utilize Pandas’ integration with Matplotlib to create a bar chart visualizing the average salaries for each experience level. Ensure the chart includes appropriate labels, a title, and a color scheme that enhances readability.

```python
# start figure
plt.figure(figsize=(10, 6))

plt.bar(grouped_salaries['experience_level'], grouped_salaries['mean'], color='#1f77b4', alpha=0.8)

# labels etc.....
plt.xlabel("experience level", fontsize=14)
plt.ylabel("average salary in USD", fontsize=14)
plt.title("Average Salary by Experience Level in High-Tech (circa 2020)", fontsize=16)
plt.xticks(rotation=45)

# show the plot
plt.tight_layout()  # no cut labels
plt.show()
```

```python
# group by 'experience_level' then aggregates
grouped_salaries_job = df_cleaned.groupby(['job_title', 'experience_level'])['salary_in_usd'].agg(['mean', 'median']).reset_index() # reset_index to plot!!!!!!!!
grouped_salaries_job = grouped_salaries_job.sort_values(by='job_title')

grouped_salaries_job
```

```python
mid_exp = grouped_salaries_job[grouped_salaries_job['experience_level'] == 'MI'].sort_values('mean')

# start figure
plt.figure(figsize=(10, 6))

plt.bar(mid_exp['job_title'], mid_exp['mean'], color='#1f77b4', alpha=0.8)

# labels etc.....
plt.xlabel("job title", fontsize=14)
plt.ylabel("average salary in USD", fontsize=14)
plt.title("Average Mid Lvl Salary by Job Title in High-Tech (circa 2020)", fontsize=16)
plt.xticks(rotation=90)

# show the plot
plt.tight_layout()  # no cut labels
plt.show()
```

```python
ent_exp = grouped_salaries_job[grouped_salaries_job['experience_level'] == 'EN'].sort_values('mean')

# start figure
plt.figure(figsize=(10, 6))

plt.bar(ent_exp['job_title'], ent_exp['mean'], color='#1f77b4', alpha=0.8)

# labels etc.....
plt.xlabel("job title", fontsize=14)
plt.ylabel("average salary in USD", fontsize=14)
plt.title("Average Entrance Lvl Salary by Job Title in High-Tech (circa 2020)", fontsize=16)
plt.xticks(rotation=90)

# show the plot
plt.tight_layout()  # no cut labels
plt.show()
```
