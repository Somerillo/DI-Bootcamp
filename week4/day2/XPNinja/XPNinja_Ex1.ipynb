## Background

The Titanic dataset is a popular dataset used in data analysis and machine learning. It contains information about passengers who were on the RMS Titanic, including whether they survived, their age, ticket class, fare, and more. This exercise is an exploration of this dataset to uncover insights and patterns.

## Dataset Overview

The Titanic dataset typically includes the following columns:

- PassengerId: A unique identifier for each passenger.
- Survived: Indicates if a passenger survived (1) or did not survive (0).
- Pclass: Ticket class (1st, 2nd, or 3rd class).
- Name: Passenger’s name.
- Sex: Passenger’s sex.
- Age: Passenger’s age.
- SibSp: Number of siblings/spouses aboard.
- Parch: Number of parents/children aboard.
- Ticket: Ticket number.
- Fare: Passenger fare.
- Cabin: Cabin number.
- Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

## 4. Exploratory Analysis

- Univariate Analysis: Examine individual variables. Plot histograms or bar charts to see the distribution of ages, fare, etc.
- Bivariate Analysis: Explore relationships between pairs of variables. Investigate if there’s a relationship between survival and ticket class or age and survival.
- Multivariate Analysis: Investigate more complex relationships involving three or more variables.

```python
df.columns
```

```python
# define cols to plot
cols = ['Age', 'Fare', 'Sex', 'sibsp', 'Pclass', '2urvived']
```

```python
# pairplot to visualize the relationships between the features
sns.pairplot(df[cols], hue='2urvived', diag_kind='kde')
plt.show()
```

On the age histogram there is a big peak in age around 30. Its distribution is a bimodal, the smoller mode corresponds to the children.
Refering on the `sex` variable, the label 1 has a very similar survival. The label 0 (possibly men) has the biggest amount of deaths.
The `Pclass` variable shows a similar survival rate (approx 1/3) for 1st and 2nd class. The 3rd class presents a similar amount of survivors but also presents the biggest amount of deaths.

```python
# histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

There are three different modes in this histogram. The first is due the children and is the smaller distribution on the left side, it is right skewed. The second mode (around 22) is due the passengers in general, also right-skewed. Finally there is a very defined peak (almost a delta) just before 30 years, very possible due the cabin personel.

```python
# Heatmap of survival rates by class and sex
survival_rate = df.groupby(['Pclass', 'Sex'])['2urvived'].mean().unstack()
sns.heatmap(survival_rate, annot=True, fmt='.2%', cmap='coolwarm')
```

For women (sex = 1), more than 60% survived when in 1st and 2nd class. Their survival rate falls to half of it for 3rd class.
Only a fourth of men survived when in 1st class. Second and third class men survived below 10%.

```python
# Bin ages into groups
df['age_group'] = pd.cut(df['Age'], bins=[0, 10, 30, 80])

# Convert age_group to string for plotting
df['age_group'] = df['age_group'].astype(str)

# FacetGrid for multivariate relationships
g = sns.FacetGrid(df, col='Pclass', row='Sex', hue='2urvived')
g.map(sns.histplot, 'age_group', multiple='stack')
```

```python
# Scatter plot with hue for survival
sns.scatterplot(
    data=df, x='Age', y='Fare', 
    hue='2urvived', style='Pclass', 
    palette={0: 'red', 1: 'green'}
)
```
