in the no_of_votes field we see that std > mean, there is a big spread. Meaning the amount of voters vary greatly from film to film, the difference from the least to most voted film is of three magnitude orders.

# 2. Data Preprocessing:

    * Handle missing values and convert data types as needed.

'Handling missing values' is a wide therm without a clear purpouse, is it for a ML model?, is for predictions and how precise they need to be?, even more before the statistical analysis could be sometimes not so good?

Anyway... we can try a general approach:

```python
# so we can handle df as we please
df = df.copy()

# flag rows just in case
df['Certificate_missing'] = df['Certificate'].isnull()
df['Meta_score_missing'] = df['Meta_score'].isnull()
df['Gross_missing'] = df['Gross'].isnull()

# impute certificate missing values with mode (categorical)
certificate_mode = df['Certificate'].mode()[0] if not df['Certificate'].mode().empty else None # need mode[0] because can be more than one mode
df['Certificate'] = df['Certificate'].fillna(certificate_mode)

# impute meta score missing values with mean (numerical)
meta_score_mean = df['Meta_score'].mean()
df['Meta_score'] = df['Meta_score'].fillna(meta_score_mean)

# the gross field is in str format, so we need to handle a bit more
df["Gross"] = df["Gross"].str.replace(",", "", regex=False)  # remove commas
df["Gross"] = pd.to_numeric(df["Gross"], errors='coerce')  # to numeric, handling errors
gross_mean = df['Gross'].mean(skipna=True)  # calculate mean, skip NaNs
df["Gross"] = df["Gross"].fillna(gross_mean).astype(int)  # impute mean --> convert to int


# checking...
print(f'remain a total of {df.isnull().sum().sum()} missing values')
```

```python
# fix release year
df = df[df['Released_Year'].str.isnumeric()] # keep only numeric entries
df['Released_Year'] = df['Released_Year'].astype(int)
```

```python
# fix runtime
df['Runtime'] = df['Runtime'].str.extract(r'(\d+)') # raw string for regex
df['Runtime'] = df['Runtime'].dropna() # drop nan if any
df['Runtime'] = df['Runtime'].astype(int) # finally numeric
```

```python
df.info()
```

## 3. Rating Trends Over Years:

    * Create a line plot with Matplotlib showing trends in IMDB_Rating over the years.



```python
plt.figure(figsize=(10, 5))

# to plot need to sort the dataframe by year
df = df.sort_values(by='Released_Year', ascending=True)
plt.plot(df['Released_Year'], df['IMDB_Rating'])

# Adding titles and labels
plt.title('trends in IMDB_Rating over the years')
plt.xlabel('')
plt.ylabel('IMDB Rating')

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

# 4. Genre Popularity Analysis:

    * Generate a Seaborn bar plot to compare the number of movies across different genres.



```python
# need to count movies in each genre to generate the columns
count_genre = df['Genre'].value_counts().reset_index()
display(count_genre)
```

There are mixed genres, that is not ok to analyze individualy...

```python
# create empty dataframe, and allow manipulation
count_genre = pd.DataFrame().copy()

# split the mixed genres into lists
count_genre['Genre'] = df['Genre'].str.split(', ')

# explode the lists into separated rows
exploded_genres = count_genre.explode('Genre')

# finally count occurrences of each unique genre
genre_counts = exploded_genres['Genre'].value_counts().reset_index()

display(genre_counts)
```

Now we have only 20 individual genres and the count skyrocketed

We proceed to plotting:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='Genre', data=genre_counts, hue='Genre', palette='viridis')

plt.title('nobody reads the titles')
plt.xlabel('# of movies')
plt.ylabel('')

plt.show()
```

Life is full of drama and also the movies

## 5. Directorâ€™s Impact on Ratings:

* Use a Matplotlib scatter plot to examine the relationship between directors and movie ratings.



```python
# ['Director', 'IMDB_Rating', 'Meta_score']


plt.figure(figsize=(10, 6))

plt.scatter(df['Director'], df['IMDB_Rating'])

plt.title('scatter plot to examine the relationship between directors and movie ratings')
plt.xlabel('director')
plt.ylabel('IMDB_Rating')

plt.xticks(rotation=45)

plt.show()
```

Impossible to read, even titlting labels

```python
plt.figure(figsize=(10, 6))

plt.scatter(df['Director'], df['Meta_score'])

plt.title('scatter plot to examine the relationship between directors and Meta_score')
plt.xlabel('director')
plt.ylabel('IMDB_Rating')

plt.xticks(rotation=45)

plt.show()
```

This is why i dont like to do value imput... anyway the meta score should be without flagged values if we want to plot/study tendencies

Now for the director vs rating plot... we can make individual directors, maybe even selecting from a top something

```python
grouped_df = df.groupby('Director').agg(
    Average_IMDB_Rating=('IMDB_Rating', 'mean'),
    Max_Meta_Score=('Meta_score', 'max')
).reset_index()

print('## imdb avg score:')
display(grouped_df.sort_values(by='Average_IMDB_Rating', ascending=False).head(10))

print('\n## meta score max score:')
display(grouped_df.sort_values(by='Max_Meta_Score', ascending=False).head(15))
```

```python
top_directors = grouped_df[grouped_df['Max_Meta_Score'] >= 100]['Director'].tolist()
# top_directors
```

```python
# filter df for top directors according meta score
filtered_df = df[df['Director'].isin(top_directors)]


plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Director'], filtered_df['IMDB_Rating'])

plt.title('scatter plot to examine the relationship between directors and movie ratings')
plt.xlabel('')
plt.ylabel('IMDB Rating')

plt.xticks(rotation=45)

plt.show()
```

The most consagrated director by the critics and at the same time valorated by imdb rating is Ford Coppola (twice)

## 7. Box Plot of Genres vs. Ratings:

* Visualize the distribution of IMDB_Rating across different Genres using a Seaborn box plot.



```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='IMDB_Rating', data=df)

plt.title('Distribution of IMDB Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating')

plt.show()
```

## 8. Correlation Heatmap:

* Create a heatmap using Seaborn to identify correlations between numerical columns like IMDB_Rating, Meta_score, No_of_votes, and Gross.



```python
# frist create correlation matrix
correlation_matrix = df[['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']].corr()

# then plot
plt.figure(figsize=(10, 8))  # Set figure size
sns.heatmap(correlation_matrix, annot=True, cmap='rocket', fmt='.2f', linewidths=0.5)

plt.title('Correlation Heatmap')
plt.show()
```

No_of_Votes is correlated with imdb_rating and gross, interestingly meta score is only correlated (and midly low) with imdb rating
