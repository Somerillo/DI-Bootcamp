```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
# numeric_columns = ['Age', 'Annual_Income_(k$)', 'Spending_Score']

# totals = df[numeric_columns].sum(axis=0)
# df['Annual_Income_normalized'] = df['Annual_Income_(k$)'] / totals['Annual_Income_(k$)']
# df['Spending_Score_normalized'] = df['Spending_Score'] / totals['Spending_Score']

# print(df.head())


plt.figure(figsize=(8, 6))

bins = round(len(df) ** 0.5)
sns.histplot(x='Annual_Income_(k$)', data=df, kde=True, color='skyblue', bins=bins, stat='density')

plt.title('some title')
plt.show()
```

```python
# Create a KDE Plot for Age: Generate a KDE Plot for the Age column to examine the distribution of customer ages.
# Set the bandwidth for the KDE to 0.5 for finer granularity.
# Use the fill=True parameter to shade the area under the KDE curve.

sns.histplot(df['Age'], kde=True, kde_kws={'bw_adjust': 0.5}, fill=False, stat='density')
```

```python
plot = sns.jointplot(data=df, x='Spending_Score', y='Age')
```

```python
plot = sns.JointGrid(data=df, x='Spending_Score', y='Age')
plot.plot_joint(plt.contour)
plot.plot_marginals(sns.histplot)
```
