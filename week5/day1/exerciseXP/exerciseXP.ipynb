# üåü Exercise 1 : Understanding Data Visualization
Instructions

    Explain why data visualization is important in data analysis.
    Describe the purpose of a line graph in data visualization.




Data visualization is a very important part of DA because it simplifies the information, and allows to visualize trends, patterns, correlations, etc.

A line graph is maybe the second most basic type of graph (preceded only by scatter plot). A 2-D scatter plot is a direct Cartessian representation of a y(x) function. Being 'y' the dependent variable and 'x' the independent variable, i.e. for each value that takes x there is a corresponding value for y. A point in the scatter plot is determined then by its (x, y) coordinates. For a set of (x1, x2, ...) we have their corresponding (y1, y2, ...), and when we connect the coordinates (x1, y1), (x2, y2), ... we have a line plot.

The line plot allows to have a visual representation of the function behaviour, like temporal evolution, trending, correlation, etc.

Sometimes the y(x) dependency can be associated/intrapolated/extrapolated with a specific equation like *y = ax + b + error*, *y = 
A sin(x) + error*, etc. Which allows to give a behavioural law and/or make predictions.

# üåü Exercise 2 : Creating a Line Plot for Temperature Variation
Instructions

    Create a simple line plot using Matplotlib that represents the temperature variations over a week.
    Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.


```python
from matplotlib import pyplot as plt

day = ['Su', 'Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa']
eg = [72, 74, 76, 80, 82, 78, 75]

# instantiate figure
plt.figure(figsize=(14, 3))

# define function
plt.plot(day, eg)

# labels
plt.xlabel('Day')
plt.ylabel('Temperature (¬∞F)')
plt.grid(axis='y', alpha=0.4)

# plot the chart
plt.tight_layout()
plt.show()
```

# üåü Exercise 3 : Visualizing Monthly Sales with a Bar Chart
Instructions

    Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.
    Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]) and label the x-axis as ‚ÄúMonth‚Äù and the y-axis as ‚ÄúSales Amount ($)‚Äù.

# üåü Exercise 4 : Data Visualisation
Instructions

Using the dataset: sales_data.csv

    Calculate the total quantity of products sold by the company.
    Identify the category that had the highest revenue and calculate how much revenue it generated.
    Determine the average revenue per sale in the dataset.
    Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).


```python
import pandas as pd

path = 'sales_data.csv'
df = pd.read_csv(path)
print(df.head())
```

```python
print(f'total quantity is: {df['quantity'].sum():.0f}')
```

```python
df_grouped = df.groupby('category')['revenue'].sum()

max_rev = df_grouped.max()
max_cat = df_grouped.idxmax()

print(f'category: {max_cat}, revenue: {max_rev}')
```

```python
tot_revenue = df['revenue'].sum()
tot_sales = len(df)

print(f'avg revenue per ssale is {tot_revenue/tot_sales:.1f}')
```

```python
"""
When we convert to period, we need to be careful about the output format!!!!!
"""

# ensure date is in format
df['date'] = pd.to_datetime(df['date'])

# create a quarter column
df['quarter'] = df['date'].dt.to_period('Q') # OJO CON EL FORMATO!!!!!!!!!!!

# group by quarter
quarterly_revenue = df.groupby('quarter')['revenue'].sum().reset_index()

# Convert 'quarter' to string for compatibility with Matplotlib
quarterly_revenue['quarter'] = quarterly_revenue['quarter'].astype(str)

# plot it
plt.figure(figsize=(12, 5))
plt.bar(quarterly_revenue['quarter'], quarterly_revenue['revenue'])
plt.xlabel('quarter')
plt.ylabel('total revenue')
plt.ylim(bottom=180000, top=200000)
plt.show()
```

# üåü Exercise 5 : Data Visualisation Using MatPlotLib
Instructions

    Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
    Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
    Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.


```python
import numpy as np

# function dominion
x = np.linspace(-10, 10, 111)

y = x ** 2

plt.figure(figsize=(10, 8))
plt.plot(x, y)
plt.title('a title')
plt.xlabel('x')
plt.ylabel('y=x^2')
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure(figsize=(10, 7))
plt.bar(products, sales)
plt.title('appropriately')
plt.xlabel('products')
plt.ylabel('sales')
plt.show()
```

```python
# data
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
colors = ['green', 'yellow', 'purple', 'brown']

# pie chart
plt.figure(figsize=(7,7))
plt.pie(sizes, labels=fruits, colors=colors, autopct='%1.f%%', startangle=90)

# legends
plt.title('fruit colors')
plt.legend(fruits)
plt.show()
```
