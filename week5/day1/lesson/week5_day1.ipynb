```python
import pandas as pd

df = pd.DataFrame({'Values':[10,20,30,40,50,60,70]})

df['Normalized_1_to_10']= 1+9*(df['Values']-df['Values'].min())/(df["Values"].max()-df["Values"].min())
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler(feature_range=(1,10))
df['Normalized_1_to_10_minmax'] = scaler.fit_transform(df[['Values']])

df
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()
df['Normalized_1_to_10_standard'] = scaler.fit_transform(df[['Values']])

df
```

```python
#Create a line plot for the function y = x^2 for x ranging from 0 to 10. x = 0, 1, 2...10 y = 0, 1, 4, .... 100

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11)
y = x**2

filter_indices = y<=50
x = x[filter_indices]
y = y[filter_indices]

plt.figure()
plt.plot(x,y, marker='o', color='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.title('Line Plot')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.arange(0, 11)  # x values: 0, 1, 2, ..., 10

# Calculate y values for y = x^2
y = x**2

# Create the plot
plt.figure(figsize=(8, 5))

# Scatter plot for the points
plt.plot(x, y, marker = "o",color='red',linestyle='-', label='Scatter Points', zorder=2)



# Add title, labels, grid, and legend
plt.title('Scatter Plot with Line for y = x^2', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```

```python
x = np.arange(0,11)

y1 = x**2
y2 = x**4

fig,ax1 = plt.subplots(figsize =(8,5))

ax1.plot(x,y1,marker='o', color = "blue", label="y=x^2")
ax1.set_xlabel("x")
ax1.set_ylabel("y1", color="blue")
ax1.grid(True, linestyle='--', alpha=0.9)
ax1.legend()

ax2 = ax1.twinx()
ax2.plot(x,y2,marker='o', color = "red", label="y=x^4")
ax2.text(x[2],y2[2], 'important point',color = "green")
ax2.set_ylabel("y2", color="red")


fig.suptitle("Comparison of y=x^2 and y=x^4")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate x values
x = np.arange(0, 11)  # x values: 0, 1, 2, ..., 10

# Calculate y and z values for y = x^2 and z = x^4
y = x**2
z = x**4

# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the data in 3D space
ax.plot(x, y, z, label='3D curve: y = x^2, z = x^4', marker='o', color='purple')

# Add labels and title
ax.set_title('3D Plot of y = x^2 and z = x^4', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y = x^2', fontsize=12)
ax.set_zlabel('z = x^4', fontsize=12)
ax.legend()

plt.show()
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the data in 3D space
ax.plot(x, y, z, label='3D curve: y = x^2, z = x^4', marker='o', color='purple')

# Add labels and title
ax.set_title('3D Plot of y = x^2 and z = x^4 with Annotation', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y = x^2', fontsize=12)
ax.set_zlabel('z = x^4', fontsize=12)

# Annotate a point
mid_index = len(x) // 2  # Choose a point in the middle
ax.scatter(x[mid_index], y[mid_index], z[mid_index], color='green', s=150, label='Annotated Point')
ax.text(x[mid_index]-1, y[mid_index], z[mid_index]+1,
        f'({x[mid_index]}, {y[mid_index]}, {z[mid_index]})',
        color='red', fontsize=10)

# Add legend
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('sample_data/california_housing_train.csv')

# 1. Display the first 10 rows of the dataset
first_10_rows = df.head(10)
print("First 10 rows of the dataset:")
print(first_10_rows)

# 2. Show all streets (rows) with a population over 800
over_800_population = df[df['population'] > 800]
print("\nStreets with population over 800:")
print(over_800_population)

# 3. Graph the population distribution
plt.figure(figsize=(10, 6))
plt.hist(df['population'], bins=100, color='blue', alpha=0.7, edgecolor='black')
plt.title('Distribution of Population', fontsize=14)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(alpha=0.5)
plt.show()

# 4. Remove streets with few houses but a high price
# Define few houses as "households <= 10" and high price as "median_house_value > 500000"
df_filtered = df[~((df['households'] <= 500) | (df['median_house_value'] > 499000))]

# 5. Graph the remaining streets' median_housing_value based on population
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['population'], df_filtered['median_house_value'], alpha=0.5, color='green')
plt.title('Median Housing Value vs Population (Filtered)', fontsize=14)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Median Housing Value', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
