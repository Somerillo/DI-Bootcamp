## World Happiness Report 2019 Dataset Column Descriptions

- **Overall rank**: This indicates the position of a country in the happiness ranking, with 1 being the happiest.
  
- **Country or region**: The name of the country or region being evaluated.

- **Score**: A numerical representation of happiness on a scale from 0 to 10, based on survey responses. This score is derived from the Cantril ladder method, where respondents rate their current lives on a scale from 0 (worst possible life) to 10 (best possible life).

- **GDP per capita**: This reflects the economic production per person in each country, adjusted for purchasing power parity. It serves as an indicator of economic well-being.

- **Social support**: This measures the perceived availability of social support, indicating how much individuals feel they have someone to count on in times of trouble.

- **Healthy life expectancy**: The average number of years individuals are expected to live in good health, which is a significant factor influencing happiness and well-being.

- **Freedom to make life choices**: This reflects how much freedom individuals feel they have in making choices about their lives, which can significantly impact overall happiness.

- **Generosity**: This measures the extent of charitable giving and altruism within a society, often assessed through survey questions about donations and volunteer work.

- **Perceptions of corruption**: This assesses how citizens perceive corruption in their government and business sectors, which can affect trust and satisfaction with life.



**References**: [World Happiness Reports](https://worldhappiness.report/news/media-round-up-on-the-2019-world-happiness-report/)

## Social Support and Happiness

```python
# instantiate figure
plt.figure(figsize=(10, 8))

# create figure
plt.scatter(df['Social support'], df['Score'], label='data points')

# linear regression model
X = df[['Social support']]
y = df['Score']
model = LinearRegression()
model.fit(X, y)

# predict
y_pred = model.predict(X)

# Calculate slope (m) and intercept (b)
slope = model.coef_[0]
intercept = model.intercept_
# Calculate R-squared value
r_squared = model.score(X, y)
# Create trend line equation string
equation_str = f"y = {slope:.2f}x + {intercept:.2f}\n$R^2$ = {r_squared:.2f}"

# plot the trend line
plt.plot(df['Social support'], y_pred, color='red', label='trend line')

# labels
plt.title('Happiness Score vs Social Support')
plt.xlabel('Social support')
plt.ylabel('Score')
plt.legend(title=equation_str) # add equation to legend

plt.show()
```

We can appreciate some (not great, not terrible) linear trend between the happiness score and the social support. We appreciate a positive correlation, althoug there is some scattering in the chart and the linear regression model only explains the 60% of the points (R=0.6), which is not that bad.

So we can infer while social support is an important factor influencing happiness, it is not the sole determinant.

```python
df_2016 = pd.read_csv('archive/2016.csv')
df_2016.head()
```

```python
# create mapping from 2016 df
region_maping = dict(zip(df_2016['Country'], df_2016['Region']))

# map the regions to the new column in df
df['Region'] = df['Country or region'].map(region_maping)
print(f'there are {df['Region'].isnull().sum()} null values in the region column of df')
```

```python
df_no_region = df[df['Region'].isnull()].copy()
display(df_no_region)
```

We have now a df of countries we couldnt assingn a region, we can input it manually or remove the from our analysis. And we`ll opt for the 2nd option :)

```python
df_clean = df[~df['Region'].isnull()].copy()
df_clean = df_clean.rename(columns={'Country or region': 'Country'})
print("### Dataset Information")
df_clean.info()
print("\n### Dataset values")
df_clean['Region'].value_counts()
```

```python
# # we need to break this dataframe:
# plt.scatter(df_clean['GDP per capita'], df_clean['Healthy life expectancy'])

# common scale for all subplots
upper_gdp = df_clean['GDP per capita'].max()
lower_gdp = df_clean['GDP per capita'].min()
upper_healthy = df_clean['Healthy life expectancy'].max()
lower_healthy = df_clean['Healthy life expectancy'].min()

# need UNIQUE regions
regions = df_clean['Region'].unique()

# need setup number of subplots
num_regions = len(regions)
cols = 2
rows = (num_regions + cols - 1) // cols

# create ubplots
fig, axes = plt.subplots(rows, cols, figsize=(14, 20))
axes = axes.flatten()

# loop through region
for i, region in enumerate(regions):
    # filter df for the current region
    region_data = df_clean[df_clean['Region'] == region]

    # create scatter plot for current region
    axes[i].scatter(region_data['GDP per capita'], region_data['Healthy life expectancy'])

    # same limits for each subplot
    axes[i].set_xlim(lower_gdp, upper_gdp)
    axes[i].set_ylim(lower_healthy, upper_healthy)

    # set labels and title for subplots
    axes[i].set_xlabel('GDP per capita')
    axes[i].set_ylabel('Healthy life expectancy')
    axes[i].set_title(f'Region: {region}')
    axes[i].grid(True, alpha=.3)


# Hide any unused subplots (if any)
for j in range(num_regions, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# sample data
gdp_mean = df_clean.groupby('Region')['GDP per capita'].mean()
healthy_mean = df_clean.groupby('Region')['Healthy life expectancy'].mean()

# instantiate figure
fig, ax1 = plt.subplots(figsize=(10, 8))

# bar plot
bar_width = 0.4
x_indices = np.arange(len(regions))
bars = ax1.bar(x_indices, gdp_mean, width=bar_width, label='mean GDP per Capita', color='skyblue')

# bar plot labels & tit y-axis
# ax1.set_xlabel('Regions')
ax1.set_ylabel('GDP per Capita', color='#1f77b4')
ax1.tick_params(axis='y', labelcolor='#1f77b4')

# create a second y-axis for Healthy Life Expectancy
ax2 = ax1.twinx()
ax2.plot(x_indices, healthy_mean, label='Healthy Life Expectancy', color='orange', marker='o')
ax2.set_ylabel('Healthy Life Expectancy', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# x-ticks to be the region names
ax1.set_xticks(x_indices)
ax1.set_xticklabels(regions, rotation=45)

# title
plt.title('GDP per Capita and Healthy Life Expectancy by Region')


# show grid
ax1.grid(axis='y', alpha=.3)

fig.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
