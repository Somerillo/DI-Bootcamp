# ðŸŒŸ Exercise 6 : Basic Hypothesis Testing
Instructions

Create a sample dataset to test the effectiveness of a new training program on employee productivity:

```python
import numpy as np

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Your task is to formulate a hypothesis regarding the training program's effectiveness
# and test it using basic statistical functions in NumPy.
```


Given a dataset, formulate a simple hypothesis and test it using basic statistical functions in NumPy.

````python
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)
````

The **hypothesis** is both performances are different, the **null hypothesis** is they are not. We stablish a probability value, typically 5% (or less, depending on the study)

````python
print("-" * 67)
print("| First we must check normality, like using a Q-Q plot.           |")
print("| This step is redundant because both arrays are np.random.normal |")
print("-" * 67)

# check both groups variance to decide A/B test method
# if variance is equal t-test, else Welch`s t-test
print(f"\n## Productivity variance before: {np.var(productivity_before)}")
print(f"## Productivity variance after: {np.var(productivity_after)}")
````

Both variances are fairly similar, and for the scope of this exercise its enough. Hence we perform a simplt t-test:

````python
t_stat, p_value = stats.ttest_ind(productivity_before, productivity_after)

print(f"## The t-statistic is: {t_stat}")
print(f"## The p-value is: {p_value}")

if p_value < 0.05:
    print(f"\n## With a p-value of {p_value}, we reject the null hypothesis")
else:
    print(f"\n## With a p-value of {p_value}, we fail to reject the null hypothesis")
````

# ðŸŒŸ Exercise 8 : Time Series Data Manipulation
Instructions

Generate time series data for the year 2023. Demonstrate slicing for the following intervals:

*    January to March
*    April to June
*    July to September
*    October to December

Generate a time series data for a specific period and demonstrate how to slice this data for different intervals.

````python
# time period of year 2023
date_range = np.arange('2023-01-01', '2024-01-01', dtype='datetime64[D]')
num_days = len(date_range)

# make time series
module = 15
constant = 16
phase = - np.pi * (0.5 + 1/12) # +1 month phi
sine = np.sin(2 * (np.pi * np.arange(num_days)/ num_days) + phase) # need another np array!!
temp_var = np.random.normal(0, 2, size=(num_days))
temp = module * sine + temp_var + constant
time_series = pd.Series(temp, index=date_range)

# plot it to check
plt.figure(figsize=(10,5))
plt.plot(time_series.index, time_series)
plt.title("Temperature in 2023")
plt.xlabel("")
plt.ylabel("deg Celcius")
plt.show()
````

````python
# now slice
jan_to_mar = time_series['2023-01-01':'2023-03-31'] # upper limit inclusive!!!
apr_to_jun = time_series['2023-04-01':'2023-06-30'] # upper limit inclusive!!!
jul_to_set = time_series['2023-07-01':'2023-09-30'] # upper limit inclusive!!!
aug_to_dec = time_series['2023-10-01':'2023-12-31'] # upper limit inclusive!!!

# plot it as proof
plt.figure(figsize=(10,5))
plt.plot(jan_to_mar.index, jan_to_mar, label='1st quarter')
plt.plot(apr_to_jun.index, apr_to_jun, label='2nd quarter')
plt.plot(jul_to_set.index, jul_to_set, label='3rd quarter')
plt.plot(aug_to_dec.index, aug_to_dec, label='4th quarter')
plt.title("you're a wizard harry")
plt.xlabel("")
plt.ylabel("deg Celcius")
plt.legend()
plt.show()
````

# ðŸŒŸ Exercise 10 : Basic Visualization
Instructions

Use Matplotlib to visualize a simple dataset created with NumPy (e.g., a line graph of random numbers).

````python
theta = np.linspace(0, np.pi, 1000)
cosine_theta = np.cos(theta)
sine_theta = np.sin(theta)

# plot it
plt.figure(figsize=(10,5))
plt.plot(theta, cosine_theta, label="cos(x)", linewidth=3)
plt.plot(theta, sine_theta, label="sin(x)", linewidth=3)
plt.axhline(y=0, color="k", linestyle='--', linewidth=1, alpha=.6)
plt.axvline(x=0, color="k", linestyle='--', linewidth=1, alpha=.6)
plt.title("nobody reads the titles")
plt.xlabel("theta (radians)")
plt.ylabel("")
plt.legend()
plt.show()
````

````python
X = sine_theta.reshape(1, -1)
Y = sine_theta.reshape(-1, 1)

matrix = np.matmul(Y, X)

# now we plot matrix
plt.figure(figsize=(10,5))
plt.imshow(matrix, cmap='magma')
plt.title("title")
plt.colorbar() # adds the colorbar for the values
plt.show()
````
