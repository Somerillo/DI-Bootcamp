## 1. Data Preparation:

Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.

Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.
Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities_list = [f"City_{i}" for i in range (1,11)]
months_of_year = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                  "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"] # hardcoding :(

temperatures = np.random.uniform(low=-5,
                                 high=35,
                                 size=(10,12)) # the range is [-5, 35)

temperatures = np.round(temperatures, decimals=1) # remove nonsense decs

df_temp = pd.DataFrame(
    temperatures,
    index=cities_list,
    columns=months_of_year) # define df

display(df_temp)
```

## 2. Data Analysis:

Hint 1: Calculate the annual average temperature using `DataFrame.mean(axis)`.

Hint 2: Find the city with the highest and lowest average temperature using `idxmax()` and `idxmin()` methods.

Calculate the annual average temperature for each city.
Identify the city with the highest and lowest average temperature for the year.


```python
df_temp["anual_avg"] = pd.DataFrame.mean(df_temp, axis=1) # create the col for avg

df_temp["anual_avg"] = round(df_temp["anual_avg"], 1) # remove nonsense decs

display(df_temp)
```

```python
hottest_city = df_temp["anual_avg"].idxmax() # find row where max is located
coldest_city = df_temp["anual_avg"].idxmin() # find row where min is located

print("## The random extreme cities in average are:\n")
print(f"  - hottest city: {hottest_city}, with an average temp "
      f"of {df_temp.loc[hottest_city, 'anual_avg']} C")
print(f"  - coldest city: {coldest_city}, with an average temp "
      f"of {df_temp.loc[coldest_city, 'anual_avg']} C")
```

## 3. Data Visualization

```python
# instantiate figure
plt.figure(figsize=(10,6))

# plot each city row except for the avg column
for row in df_temp.index:
    plt.plot(months_of_year,
             df_temp.loc[row, months_of_year],
            #  label=f"{row}, annual avg {df_temp.loc[row,'anual_avg']}")
             label=row)

# title, labels...
plt.title("it`s not leviosa its leviosaaaa")
plt.legend()
plt.xlabel("") # redundtant
plt.ylabel("temperature (C)")
plt.show()
```

## Extra mile: temperatures distributions (to prove conclutions hypothesis)

```python
# amount of bins is the difference between top and lower temp
len_ = (35 + 5)

# one only histogram to compare all temp distribution in total
plt.figure(figsize=(10, 6))
plt.hist(df_temp.values.flatten(), bins=len_) # flatten the dataframe to create a 1D array of all values
plt.title("total temps distribution in the dataframe")
plt.xlabel("temperature")
plt.ylabel("count")

plt.show()
```
