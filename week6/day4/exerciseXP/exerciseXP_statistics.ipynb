# ðŸŒŸ Exercise 1 : Linear Regression Analysis
Instructions

Objective: Apply linear regression to a dataset and interpret the results.

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.

Dataset:
```python
    house_sizes: [50, 70, 80, 100, 120] (in square meters)
    house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)
```
Questions:

*    What is the slope and intercept of the regression line?
*    Predict the price of a house that is 90 square meters.
*    Interpret the meaning of the slope in the context of housing prices.


````python
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

# perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# the sploe and intercept of the regression line
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
````

````python
# predict price for 90 sqr m
predicted_price = slope * 90 + intercept # line eq
print(f"Predicted price for 90 sqr m: {predicted_price:.2f}")

# plot it
plt.scatter(house_sizes, house_prices)
plt.plot([0, 120], [intercept, slope * 120 + intercept], color='red')
````

It is a linear equation: the slope is the increment of y on x, and the intercept is the vertical distance to the origin.

# ðŸŒŸ Exercise 2 : Understanding ANOVA
Instructions

Objective: Test understanding of ANOVA and its application.

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

Dataset:
```python
    fertilizer_1: [5, 6, 7, 6, 5]
    fertilizer_2: [7, 8, 7, 9, 8]
    fertilizer_3: [4, 5, 4, 3, 4]
```
Questions:

*    Perform an ANOVA test on the given data. What are the F-value and P-value?
*    Based on the P-value, do the fertilizers have significantly different effects on plant growth?
*    Explain what would happen if the P-value were greater than 0.05.


````python
# Sample data from three different groups
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# Printing the results
print("F-value:", f_value)
print("p-value:", p_value)


# boxplot of the fertilizer groups
plt.boxplot([fertilizer_1, fertilizer_2, fertilizer_3], labels=['Fertilizer 1', 'Fertilizer 2', 'Fertilizer 3'])
plt.xlabel('') # redundant
plt.ylabel('Value')
plt.show()
````

With a p-value p < 0.05 we reject the null hypotesis that "all fertilizers have the same effect". And that is owr case now, so we can affirm different fertilizers have significantly different effects on plant growth.

For a p greater or equal than 0.05 we can't affirm, nor deny.



# ðŸŒŸ Exercise 3 : Probability Distributions
Instructions

*    Work with a binomial distribution: calculate probabilities for different numbers of successes.
*    Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.


````python
# parameters for the binomial distribution
n_trials = 10
p_success = 0.5 # not loaded coin

# calculate the probability of getting exactly 5 successes
k = 5
probability = stats.binom.pmf(k, n_trials, p_success)

print(f"Probability of getting {k} successes in {n_trials} trials: {probability:.4f}")
````

# ðŸŒŸ Exercise 4 : Correlation Coefficients
Instructions

  *  Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
  *  Example in code:
  ```python
  data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

````python
# i changed the dataset from the example because the correlation was practically 1
data = pd.DataFrame({'study_hours': [2,3,4,5,6,7,8,7,4,9], 'test_scores': [70,75,80,85,90,95,100,95,89,100]})

# obtain correlation coefficients
correlation_pearson, p_value_pearson = stats.pearsonr(data['study_hours'], data['test_scores'])
correlation_spearman, p_value_spearman = stats.spearmanr(data['study_hours'], data['test_scores'])

print(f"The Pearson correlation is {correlation_pearson:.2f}")
print(f"The Spearman correlation is {correlation_spearman:.2f}")
````

A strong positive correlation

# ðŸŒŸ Exercise 5 : Linear Relationships
Instructions

  *  Create a scatter plot of two variables (age and income) and visually assess the linear relationship. Use the dataset from Exercise 7.


````python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

# linear reg model
slope, intercept, r_value, p_value, std_err = stats.linregress(data['age'], data['income'])

# plot it
plt.scatter(data['age'], data['income'])
plt.plot([0, 45], [intercept, slope * 45 + intercept], color='red')
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(['Regression Line', 'Data Points'])
plt.show()

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R value: {r_value:.2f}")
print(f"P value: {p_value:.2f}")
print(f"Std err: {std_err:.2f}")
print(f"\nThe equation of the fit is: \n              y = ({slope:.2f}) x"
      f" + ({intercept:.2f})")
````

With an R^2=1 we have a perfect fit explaining 100% of the points
