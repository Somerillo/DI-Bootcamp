# Mini-project: Advanced Statistical Analysis of Apple Inc. Stock Data

## About the Data

This dataset contains stock market information for Apple Inc. (AAPL). Below is an explanation of each column:

- **Volume**: The number of shares traded during the specified time period.

- **VWAP (Volume-Weighted Average Price)**: A trading benchmark that represents the average price a security has traded at throughout the day, weighted by volume. It is calculated by dividing the total dollar value of trading in the security by the total trading volume.

- **Open**: The price at which the stock first traded when the market opened for the day.

- **Close**: The final price at which the stock traded when the market closed for the day.

- **High**: The highest price the stock reached during the trading day.

- **Low**: The lowest price the stock reached during the trading day.

- **Time**: The date and time of the recorded data point.

- **Transactions**: The number of individual trades executed during the specified time period.

This data can be used for various types of analysis, including trend identification, volatility assessment, and performance evaluation of Apple's stock over time.

For a quarterly more detailed view candlestick chart; we take the first opening value (and last for the closing one), the maximum as tthe `high` (and minimum for `lower` value), the `volume` & `transactions` are taken as the sum. So the new dataframe is a resample (not a grouping):

```python
# resample to quarterly data
df_quarterly = df.resample('QE', on='time').agg({
    'open': 'first',  # use the 1st
    'high': 'max', # use the quarterly max
    'low': 'min', # use the quarterly min
    'close': 'last', # use the last closing
    'volume': 'sum', # the total quarter one
    'transactions': 'sum' # the total quarter one
}).copy() # just in case use always copy

# make the quarterly candlestick chart
mpf.plot(df_quarterly, type='candle',
         style='charles', # after this we need to restore matplotlib
         title='Quarterly Candlestick Chart',
         ylabel='Price',
         volume=False, # we'll avoid seeing the volume for the moment
         figsize=(16, 9))
plt.show()

# we restore matplotlib to its default values
plt.rcParams.update(plt.rcParamsDefault)
```

In the candlestick chart we observe a significant drop in Apple's stock price in the quarters following March 31, 2022. That is during the general tech sector downturn: the technology sector as a whole experienced a significant correction during this period, [with Apple's stock following the broader trend ](https://www.forbes.com/sites/qai/2022/10/05/why-is-apple-stock-down-right-now/). This trend is reverted later by two highly growing quarters.

The transactions volume chart is somewhat different from the closing prices:

```python
# make the quarterly volume candlestick chart
mpf.plot(df_quarterly, type='ohlc',
         style='charles',  # after this we need to restore matplotlib
         title='Quarterly Volume Chart',
         ylabel='Volume', # for the main panel (volume in this case)
         ylabel_lower='Price', # for the lower panel (price in this case)
         volume=True,
         show_nontrading=True,
         figsize=(16, 9),
         panel_ratios=(1,1),
         main_panel=1,
         volume_panel=0)
plt.show()

# we restore matplotlib to its default values
plt.rcParams.update(plt.rcParamsDefault)
```

The total transaction volume by year plot is a mirror image of the average closing price by year plot. The volume (number of shares traded) decreased in a pattern inverse to the closing price increase. As Apple's stock price has risen over the years, the trading volume has tended to decrease.

## 3. Statistical Analysis

The return on an investment, often called the Rate of Return (RoR), is a measure of the profit or loss generated relative to the amount invested. It is typically expressed as a percentage. Here's how to calculate it mathematically:

RoR = [(Current Value - Initial Value) / Initial Value] × 100

Where:
- Current Value: The final value of the investment
- Initial Value: The original amount invested

The `pct_change()` method in pandas is a convenient and efficient way to calculate percentage changes between consecutive values in a DataFrame or Series:

```python
# calculate daily returns using percentage change
df_returns = pd.DataFrame()
df_returns['time'] = df['time'].copy()
df_returns['return'] = df['close'].copy().pct_change().dropna()

# get the descriptive statistics
mean_return = df_returns['return'].mean()
std_dev_return = df_returns['return'].std()
skewness = df_returns['return'].skew()
kurtosis = df_returns['return'].kurtosis()

# print the results
print(f"Mean Return: {mean_return}")
print(f"Standard Deviation of Returns: {std_dev_return}")
print(f"Skewness of Returns: {skewness}")
print(f"Kurtosis of Returns: {kurtosis}")
```

While the stock has a slight positive average return, it also exhibits notable volatility and risk characteristics. The negative skewness indicates a tendency for occasional larger losses, while high kurtosis highlights the presence of extreme values in the return distribution. Potential for both significant gains and losses in this stock's performance.

```python
# figure with 1 row and 2 plots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# histogram of returns
sns.histplot(df_returns['return'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Returns')
axes[0].set_xlabel('Return')
axes[0].set_ylabel('Frequency')

# q-q plot for dist comparison
returns = df_returns['return'].dropna()
(quantiles, values), (slope, intercept, r) = stats.probplot(returns)

axes[1].scatter(quantiles, values)
axes[1].plot(quantiles, slope * quantiles + intercept, color='red', lw=2)
axes[1].set_title('Q-Q Plot of Returns')
axes[1].set_xlabel('Theoretical Quantiles')
axes[1].set_ylabel('Sample Quantiles')


# adjust spacing between layouts
plt.tight_layout()
plt.show()
```

The Q-Q plot shows a heavy tailed distribution. Meaning that while most of the returns may cluster around the mean, there is a substantial risk of extreme losses or gains (as it was previously mentioned). Also the line is almost horizontal, indicating the returns are more concentrated around the mean than expected for a normal distribution.

Computing rolling statistics, such as the 30-day moving average and standard deviation, helps us to smooth out short-term fluctuations and highlight longer-term trends in financial data.

```python
# calculate moving averages for different time periods
short_window = 30 # monthly
long_window = 120 # quarterly
df['Short_MA'] = df['close'].rolling(window=short_window).mean()
df['Long_MA'] = df['close'].rolling(window=long_window).mean()

# plot the windows
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['close'], label='Closing Price', alpha=0.5)
plt.plot(df['time'], df['Short_MA'], label=f'Short-Term MA ({short_window} days)', linewidth=2)
plt.plot(df['time'], df['Long_MA'], label=f'Long-Term MA ({long_window} days)', linewidth=2)
plt.title('Closing Price and Moving Averages')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()



# calculate correlations between moving averages and volume
correlation_short_volume = np.corrcoef(df['Short_MA'][long_window:], df['volume'][long_window:])[0, 1]
correlation_long_volume = np.corrcoef(df['Long_MA'][long_window:], df['volume'][long_window:])[0, 1]

print(f"Correlation between short-term moving average and volume: {correlation_short_volume}")
print(f"Correlation between long-term moving average and volume: {correlation_long_volume}")

# calculate correlations between the moving averages themselves
correlation_ma = np.corrcoef(df['Short_MA'][long_window:], df['Long_MA'][long_window:])[0,1]
print(f"Correlation between short and long term moving averages: {correlation_ma}")
```

From the moving averages plot, after comparing for windows of 30 and 120 days, we find a medium and consistent correlation of -0.51 between the closing price and the number of shares (`volume`). Which confirms our observations on the candlestick charts. This relationship underscores the importance of **volume as a potential signal for price movements** in Apple stocks. The correlation between the short-term and long-term moving averages is very high at about 0.99, highlighting that both averages move in tandem.

## 4. Hypothesis Testing

For the hypothesis test we compare two periods of sustained growth with a fall in between. The stocks had a very high growth just before the [tech downturn](https://www.forbes.com/sites/qai/2023/01/19/why-were-tech-stocks-down-in-2022-and-how-long-will-the-slump-last/) around March 31, 2022. After the stocks fall we see a sustained recovery growth, the hypotheses are then:
- Null Hypothesis (H0): There is no significant difference in the average returns before the downturn period (2021-03-31 to 2022-03-31) and during the recovery period (2022-09-30 to 2023-09-30).
- Alternative Hypothesis (H1): There is a significant difference in the average daily returns between the downturn period and the recovering period.

In simpler words, the hypothesis test is to check if the recovery is as before the fall.

Notice, another important growth for the tech sector during the COVID pandemic.



```python
# define the date range
start_date = '2019-05-01'
end_date = '2023-07-12' # last day in dataset

# slice the dataframe
df_sliced = df_quarterly[(df_quarterly.index >= start_date) & (df_quarterly.index <= end_date)].copy() # always copy just in case
# df_sliced = df[(df['time'] >= start_date) & (df['time'] <= end_date)].copy()
# df_sliced.set_index('time', inplace=True)

# sort the index to ensure it's in chronological order
df_sliced.sort_index(inplace=True)

# define plot mark before downturn period
rect_start = pd.to_datetime('2021-07-01')
rect_end = pd.to_datetime('2021-12-31')

# define plot mark after downturn period
rect_start2 = pd.to_datetime('2023-01-01')
rect_end2 = pd.to_datetime('2023-06-30')

# define plot mark on covid period
rect_start3 = pd.to_datetime('2020-04-01')
rect_end3 = pd.to_datetime('2020-09-30')

# Create the candlestick chart
mpf.plot(df_sliced, type='candle', style='charles',
         title='Apple Stock Candlestick Chart - Comparison Periods',
         ylabel='Price',
         volume=False,
         figsize=(10, 6),
         datetime_format='%Y-%m-%d',
         fill_between=[ # highlight the periods to compare
             dict(y1=df_sliced['low'].min(),
                  y2=df_sliced['high'].max(),
                  where=(df_sliced.index >= rect_start) & (df_sliced.index <= rect_end),
                  alpha=0.4,
                  color='lightgreen'),  # first grow
             dict(y1=df_sliced['low'].min(),
                  y2=df_sliced['high'].max(),
                  where=(df_sliced.index >= rect_start2) & (df_sliced.index <= rect_end2),
                  alpha=0.4,
                  color='lightgreen'),  # recovery period
             dict(y1=df_sliced['low'].min(),
                  y2=df_sliced['high'].max(),
                  where=(df_sliced.index >= rect_start3) & (df_sliced.index <= rect_end3),
                  alpha=0.4,
                  color='lightgreen')  # covid period
         ]
         )

# Show the plot
plt.show()

# we restore matplotlib to its default values
plt.rcParams.update(plt.rcParamsDefault)

# aditional info
print(f'growth during covid period: {rect_start3} - {rect_end3}')
print(f'growth before tech fall period: {rect_start} - {rect_end}')
print(f'growth for recovery period: {rect_start2} - {rect_end2}')
```

```python
# returns t-test covid vs recovery
t_stat, p_value = stats.ttest_ind(
    df_returns.loc[(df_returns['time'] >= rect_start3) & (df_returns['time'] <= rect_end3), 'return'],
    df_returns.loc[(df_returns['time'] >= rect_start2) & (df_returns['time'] <= rect_end2), 'return']
)

print(f't-test returns covid growth vs recovery growth')
print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')

# returns t-test before tech fall vs recovery
t_stat, p_value = stats.ttest_ind(
    df_returns.loc[(df_returns['time'] >= rect_start) & (df_returns['time'] <= rect_end), 'return'],
    df_returns.loc[(df_returns['time'] >= rect_start2) & (df_returns['time'] <= rect_end2), 'return']
)

print(f'\nt-test returns before tech fall growth vs recovery growth')
print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')
```

In both cases, the p-values are well above the conventional significance level of 0.05. This indicates that there is no statistically significant difference in returns between the the COVID-19 and the final market recovery growth periods, neither the before the tech downturn and the final market recovery growth periods.

## 5. Advanced Statistical Techniques

In this section we will apply Fourier transforms to identify cyclical patterns in stock prices.

```python
# apply fourier transform
fft_vals = np.fft.fft(df['close'])
freq = np.fft.fftfreq(len(df['close']))

# identify dominant freqs
dominant_freq = freq[np.argmax(np.abs(fft_vals))]

# plot results
plt.figure(figsize=(10, 6))
plt.plot(freq, np.abs(fft_vals))
plt.axvline(x=dominant_freq, color='r',
            linestyle='--',
            linewidth=1,
            label='Dominant Frequency'
            )
plt.title('Fourier Transform of Apple Stock Prices')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.yscale('log')
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()
```

With a very strong trend and dominant frequency at 0, we can conclude that there are no observable cyclical patterns in the `close` price data.
